```python
#initialization block
from pydriller import Repository
repos = ["django", "flask", "requests"]
```

```python
target_commits = 50  # total number of commits with deleted Python lines
collected = 0

for repo_path in repos:
    print(f"Scanning repo: {repo_path}")
    
    for commit in Repository(repo_path, only_modifications_with_file_types=['.py']).traverse_commits():
        mods = getattr(commit, 'modifications', [])
        if not mods:
            continue  # skip commits with no Python modifications

        # sum deleted lines across Python files
        deleted_total = sum(mod.deleted_lines for mod in mods)
        if deleted_total == 0:
            continue  # skip commits with no deleted Python lines

        # print commit info
        print("Hash:", commit.hash)
        print("Author:", commit.author.name)
        print("Date:", commit.author_date)
        print("Message:", commit.msg)
        print("Files changed:", len(mods))
        print("Deleted lines:", deleted_total)
        print("-" * 40)

        collected += 1
        if collected >= target_commits:
            break  # stop scanning further commits in this repo

    if collected >= target_commits:
        break  # stop scanning further repos

print(f"Collected {collected} commits with deleted Python lines.")
```
